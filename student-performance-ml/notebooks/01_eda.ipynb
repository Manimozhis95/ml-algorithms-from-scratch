```python
#Distribution
import seaborn as sns
import matplotlib.pyplot as plt
#sns.histplot(df['math score'])
sns.histplot(df['math score'], kde=True)
plt.show()

# X‑axis → math score (0 to 100)
# Y‑axis → number of students (count)
###KDE = Kernel Density Estimate
#KDE is a smooth curve that estimates the probability density of your data.
```

observation:
1.bell‑shaped(Most students scored around the middle range)
2.Majority of students score between 50 and 80
3.slightly left‑skewed (More students score higher than very low)
4.unimodality
The distribution of math scores is unimodal, with a single peak around 60–70, indicating one dominant performance group rather than distinct subpopulations.

```python
#compare categories
sns.boxplot(x='gender',y='math score',data=df)
plt.show()
```
